### Creating a Polynomial Regression Plot

```python
# import all libraries
import numpy as np 
import matplotlib.pyplot as plt

#Create your array
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

##Create a model that fits poly regression line to the scattered Dots
mymodel = np.poly1d(np.polyfit(x, y, 3))

#Create a linespace from 1 to 22, with the highest vertical value to be 100
myline = np.linspace(1, 22)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

#### R-Squared is the relationship between the x and y-axis

##### The R-squared values ranges from 0 to 1, 0 means no correlation. while 1 means 100% correlation.

```python
##Compute R-squared value with python and sklearn
import numpy as np
from sklearn.metrics import r2_score

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = np.poly1d(np.polyfit(x, y, 3))

print(r2_score(y, mymodel(x)))
```
